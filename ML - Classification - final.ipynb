Removing outliers

```python
# Creating an outlier removal function
def outlier(dataframe_in, col_names):
    q1 = dataframe_in[col_names].quantile(0.25)
    q3 = dataframe_in[col_names].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    lower_fence  = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr
    dataframe_out = dataframe_in.loc[(dataframe_in[col_names] > lower_fence) & (dataframe_in[col_names] < upper_fence)]
    return dataframe_out
```

```python
# Removing outliers for all numerical columns except target variable 'output'
cols = ['age', 'trtbps', 'chol', 'thalachh', 'oldpeak']

dataset_3 = dataset_2
for x in cols:
    dataset_3 = outlier(dataset_3, x)
    
dataset_3
```

Checking for anomalies in categorical columns

```python
# Making a list of categorical columns
cat_cols = list(dataset_3.drop(cols, axis=1).columns)

# Printing values of categorical columns to check for anomalies
for x in cat_cols:
    print(x)
    print(dataset_3[x].value_counts())
    print('\n')
```

# LOGISTIC REGRESSION

```python
# Importing required packages
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import plot_roc_curve
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn import metrics
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
dataset
```

### Setting up the pipeline

Grouping features by type and storing them in lists to build the predictive model

```python
# List of features
features = list(dataset.drop("output", axis=1))

# List of categorical features
cat_features = ['sex', 'cp', 'fbs', 'restecg', 'exng', 'slp', 'caa', 'thall']

# List of numerical features
num_features = ["age","trtbps","chol","thalachh","oldpeak"]

# Target variable that should be predicted
target = ["output"]
```

Since the counts of the target/output variable are not heavily skewed (in the ratio of 1:100 or 1:1000) we will not need to conduct oversampling

```python
# Creating a copy of the dataset with only the features
X = dataset[features].copy()
```

Creating the training and testing dataset

```python
# Creating X and y training and test datasets with a train-test ratio of 0.75 and ensuring reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, dataset.output, test_size=0.25, random_state = 1)
```

Creating a pipeline to scale and encode features

```python
# constructing a numerical pipeline to scale numerical features
numerical_transformer = Pipeline(steps=[ ('imputer', SimpleImputer()), ('scaler', StandardScaler())])
```

```python
# Constructing a categorical pipeline to one-hot encode categorical features
categorical_transformer = Pipeline(steps=[ ('imputer', SimpleImputer()), ('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
# Combining the numerical and categoricaltransformer pipelines to construct a data transformer pipeline
data_transformer = ColumnTransformer(transformers=[('categorical', categorical_transformer, cat_features),
                                                   ('numerical', numerical_transformer, num_features)])
```

### Building the logistic regression model

```python
# Creating the logistic regression pipeline 
model_log = Pipeline(steps=[('data_transformer', data_transformer), ('model_log', LogisticRegression(max_iter=10000, random_state=1))])

# Training the model using train dataset
model_log.fit(X_train, y_train)

# Predicting values using test dataset
y_prediction_test_log = model_log.predict(X_test)
```

##### Evaluating the model

```python
# Classification report for logistic regression model
print(classification_report(y_test, y_prediction_test_log))
```

```python
# Using the scikit learn confusion_matrix() function to make a confusion matrix for the logstic regression model
confusion_matrix_logistic_reg = confusion_matrix(y_test, y_prediction_test_log)
confusion_matrix_logistic_reg
```

```python
# Visualizing the logistic regression confusion matrix
cm_log_display = ConfusionMatrixDisplay(confusion_matrix_logistic_reg, display_labels=['Low Chance','High Chance'])
cm_log_display.plot()
plt.show()
```

### Tuning hyperparameters with GridSearchCV

```python
# Checking parameter keys to tune the model
model_log.get_params().keys()
```

```python
# Doing a Grid Search Cross Validation on the Logistic regression
# Setting the grid search parameters
param_grid = {'data_transformer__numerical__imputer__strategy': ['mean', 'median'],
              'data_transformer__categorical__imputer__strategy': ['constant','most_frequent'],
              'model_log__penalty':['l2'],
              'model_log__solver': ['lbfgs','newton-cg','liblinear','sag','saga'],
              'model_log__max_iter': [100,200,300,400]
                                   }
               

# We use n_jobs=-1 to parallelize tasks and reduce computation time
grid_log = GridSearchCV(model_log, param_grid=param_grid, n_jobs=-1)

# Fitting the tuned model using the X and y training datasets
grid_log.fit(X_train, y_train)

# Predicting output values using test dataset
y_prediction_test_log_grid = grid_log.predict(X_test)

# Printing best hyperparameters and score
print('Best hyperparameters are: '+str(grid_log.best_params_))
print('Best score is: '+str(grid_log.best_score_))

# 20 SECONDS
```

##### Checking model performance after hyperparameter tuning using GridSearchCV

```python
# Classification report for logistic regression model after hyperparameter tuning
print(classification_report(y_test, y_prediction_test_log_grid))
```

```python
# Using the scikit learn confusion_matrix() function to make a confusion matrix for the logstic regression model
confusion_matrix_logistic_reg_grid = confusion_matrix(y_test, y_prediction_test_log_grid)
confusion_matrix_logistic_reg_grid
```

```python
# Visualizing the logistic regression confusion matrix
cm_log_display_grid = ConfusionMatrixDisplay(confusion_matrix_logistic_reg_grid, display_labels=['Low Chance','High Chance'])
cm_log_display_grid.plot()
plt.show()
```

Plotting ROC curves

```python
#Plotting the roc curves
ax = plt.gca()
metrics.RocCurveDisplay.from_estimator(model_log, X_test, y_test, ax=ax, name='Logistic Regression before tuning') 
metrics.RocCurveDisplay.from_estimator(grid_log, X_test, y_test, ax=ax, name='Logistic Regression after tuning')
plt.show()
```
